# Удаление столбцов

За удаление строк и столбцов в таблице отвечает метод drop().  
Основные параметры метода drop():  
* labels - порядковые номера или имена столбцов, которые подлежат удалению. Если их несколько, то  
передается список;  
* axis - ось совершения операции, axis=0 - удаляются строки, axis=1 - удаляются столбцы:  
* inplace - если True, происходит замена изначального DataFrame на новый, а метод ничего не возвращает; если False -  
то возвращается измененная копия DataFrame, а исходный не изменяется; по умолчанию параметр равен False

Удалим столбцы index и Coordinates. Выведем первые 5 строк таблицы.

```python
melb_df = melb_df.drop(['index', 'Coordinates'], axis=1)
melb_df.head()
```

```python
# Альтернативный вариант. Сначала восстановим melb_df
melb_df = melb_data.copy()
melb_df.drop(['index', 'Coordinates'], axis=1, inplace=True)
melb_df.head()
```

# Математические операции со столбцами

Pandas поддерживает базовые математические операции между столбцами: столбцы можно складывать, вычитать, умножать,  
делить между собой, а также возводить в степень. С помощью таких операций мы можем создавать новые признаки или  
производить операции над старыми.

Причем все операции совершаются поэлементно, очень быстро и без написания циклов.  
Быстродействие достигается за счет того, что математические операции над столбцами выполняются на языке C.

Создадим переменную total_rooms, в которой будем хранить общее количество комнат в здании. Для этого выполним  
сложение столбцов с количеством комнат, ванн и спален:

```python
total_rooms = melb_df['Rooms'] + melb_df["Bedroom"] + melb_df['Bathroom']
display(total_rooms)
```

А теперь введем прзнак MeanRoomsArea,, который соответствует средней площади одной комнаты для каждого объекта. Для этого  
разделим площадь здания на полученное ранее общее количество комнат:

```python
melb_df['MeanRoomsArea'] = melb_df['BuildingArea'] / total_rooms
display(melb_df['MeanRoomsArea'])
```

Можно ввести еще один признак - AreaRatio, коэффициент соотношения площади здания (BuildingArea) и площади участка (Landsize).  
Для этого разницу двух площадей поделим на их сумму:

```python
diff_area = melb_df['BuildingArea'] - melb_df['Landsize']
sum_area = melb_df['BuildingArea'] + melb_df['Landsize']
melb_df['AreaRatio'] = diff_area/sum_area
display(melb_df['AreaRatio'])
#display(melb_df['AreaRatio'].head())
```

Можно увидеть, что AreaRatio лежит в пределах от -1 до 1.  
Рассмотрим три случая:  
* Если рассматриваемые площади равны, то числитель зануляется и коэффициент равен 0.
* Если одна из площадей начинает доминировать над другой, то коэффициент начинает расти в отрицательную сторону, если  
площадь участка больше площади здания, и в положительную, если наоборот.  
* Если площадь здание равна 0, то коэффициент равен -1, а если площадь участка равна 0, то коэффициент равен 1.

Таким образом, AreaRatio служит своеобразным указателем соотношения площадей объекта недвижимости. Для пустырей -  
участков без строений - он будет равен -1, для домов без территории - 1, в остальных случаях можно увидеть, какая площадь больше,  
здания или участка.

```python
# Написать функцию delete_columns(df, col=[]), которая удаляет столбцы из DataFrame и возвращает новую таблицу.
# Если одного из столбцов не существует, то функция должна возвращать None
def delete_columns(df, col=[]):
    lst = list(df.keys())
    for key in col:
        if key not in lst:
            return None
    return df.drop(col, axis=1)

def delete_columns_1(df, col=[]):
    for cc in col:
        if cc not in df.columns:
            return None
    return df.drop(col, axis=1)

melb_df = melb_data.copy()
df1 = delete_columns(melb_df, ['index', 'Coordinates'])
display(melb_df.head(1))
display(df1.head(1))
```
